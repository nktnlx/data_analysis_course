## Cars Sales [mini project]

**Importing libraries**

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

**Importing dataset from a zipped csv-file in the working directory**

```python
df = pd.read_csv('cars.zip', compression='zip')

df
```

**Checking missing values and columns datatypes**

```python
df.info()
```

**Create a 'company' column with a car's manufacturer brand name**

```python
df['company'] = df['CarName'].apply(lambda x: x.split()[0])

df.company
```

**Find the number of unique company names**

```python
df.company.nunique()
```

**Check unique company names**

```python
df.company.unique()
```

**Replace typos in company names**

```python
typos = {'maxda': 'mazda',
        'Nissan': 'nissan',
        'porcshce': 'porsche',
        'toyouta': 'toyota',
        'vokswagen': 'volkswagen',
        'vw': 'volkswagen'}

for key, value in typos.items():
    df['company'] = df.company.str.replace(key, value)
```

**Find the number of unique company names now**

```python
df.company.nunique()
```

**Delete CarName and car_Id columns from the dataframe**

```python
df.drop(columns=['CarName', 'car_ID'], axis=1, inplace=True)

df
```

**Make a slice of the dataframe with predictors we'll need for regression**

```python
cars = df.loc[:, ['company', 'fueltype', 'aspiration','carbody', 'drivewheel', 'wheelbase', 'carlength','carwidth', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower', 'price']]
cars.reset_index().drop(columns='index')

cars
```

**Show correlation matrix**

```python
cars.corr()
```

**Prepare dataframe for linear regression**

```python
# finding columns with 'object' data type
cars.dtypes
```

```python
# Converting categorical variable into dummy/indicator variables.
df_dummy = pd.get_dummies(data=cars[['company', 'fueltype', 'aspiration','carbody', 'drivewheel', 'enginetype', 'cylindernumber']],
                          drop_first = True)

df_dummy
```

```python
#  concatenating df_dummy with leftovers of our dataframe
cars_lr = pd.concat([cars.drop(['company', 'fueltype', 'aspiration','carbody', 'drivewheel', 'enginetype', 'cylindernumber'], axis=1), df_dummy], axis=1)

cars_lr
```

**Make OLS regression to predict price based on horsepower**

```python
results = smf.ols('price ~ horsepower', cars_lr).fit()
print(results.summary())
```

**Now use all available predictors to predict price**

```python
X = cars_lr.drop(['price'], axis='columns') 
X = sm.add_constant(X)  
y = cars_lr['price']  

model_all = sm.OLS(y, X).fit()
print(model_all.summary())
```

**Create a regression model using all available predictors except the one linked with cars brand name**

```python
# removing columns with cars names
cols_to_keep = cars_lr.columns[~cars_lr.columns.str.startswith('company_')]
cols_to_keep
```

```python
# doing regeression
X = cars_lr[cols_to_keep].drop('price', axis='columns')
X = sm.add_constant(X)
y = cars_lr['price']

model_no_company = sm.OLS(y, X).fit()
print(model_no_company.summary())
```

The chosen model explains approximately 90% of the variance. Among the predictors, 10 out of 27 were not significant (p> 0.05).   
For a single change in horsepower, the price rises by 86.8164 USD.

**Showing a list of statistically significant predictors**

```python
results_summary = model_no_company.summary()
results_as_html = results_summary.tables[1].as_html()
s = pd.read_html(results_as_html, header=0, index_col=0)[0]

s[s['P>|t|'] < 0.05]
```
